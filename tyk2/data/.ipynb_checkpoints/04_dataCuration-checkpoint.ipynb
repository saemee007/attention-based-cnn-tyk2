```python
import pandas as pd
from tqdm import tqdm
import numpy as np
from sklearn.model_selection import train_test_split
```

### float자리에 dict있는 데이터 제거

```python
data = pd.read_json('tyk2.json', lines=True)
data
```

```python
err_idx = []
for i in tqdm(range(len(data))):
    try: 
        data.drop(columns=['smile_ft', 'id', 'subset', 'quinazoline', 'pyrimidine', 'smiles', 'active']).iloc[i,:].astype(np.float64)
    except:
        err_idx.append(i)
```

```python
err_idx
```

```python
data.select_dtypes(object).columns
```

```python
data.loc[err_idx, ['AXp-0dv', 'MDEC-33', 'Xp-0dv']]
```

### smile_ft max_len 설정

```python
data = data.drop(err_idx)
```

```python
l = []
for i in data['smile_ft']:
    l.append(len(i))
pd.Series(l).value_counts()
```

```python
max_len = 6300
for i in tqdm(range(len(data))):
    data.iat[i, 866] = data.iloc[i, 866][:max_len]
```

### Train / Test split

```python
data
```

```python
train, test = train_test_split(data, shuffle=True, test_size=0.2, random_state=42)
```

```python
train.to_json('tyk2.json',orient='records', lines=True)
test.to_json('tyk2_test.json',orient='records', lines=True)
```
